## EDA and Data Preprocessing

### Taking Out Useful Features

```python
movies = movies[['movie_id','genres','keywords','title','cast','crew']]
movies.head()
```

### Checking for Missing Values

```python
movies.isnull().sum()
```

* There are just 3 missing values , we can just **Drop** them

```python
movies.dropna(inplace=True)
movies.shape
```

### Checking "genres" Feature

```python
movies.genres[0]
```

 Here, we have to abstract **"name"** from each dictionary in genres column gor each movie.
 For this, we first have to change this object type.
* This can be done by using **literal_eval()** function of ast module.
* The **literal_eval** safely evaluate an expression node or a string containing a Python literal or container display. The string or node (a file) provided may only consist of the following Python literal structures: strings, bytes, numbers, tuples, lists, dicts, sets, booleans, and None.

```python
from ast import literal_eval
print(literal_eval(movies.genres[0]))
```

```python
def get_genres(obj):
    Genres = []
    for itr in literal_eval(obj):
        Genres.append(itr['name']) 
    return Genres 
```

```python
movies['genres'] = movies['genres'].apply(get_genres)
movies.head()
```

### Checking "keyword" Feature

```python
movies.keywords[0]
```

Similarly to genres , we have to abstract **"name"** from keywords for each movie

```python
def get_keywords(obj):
    Keywords = []
    for itr in literal_eval(obj):
        Keywords.append(itr['name']) 
    return Keywords 
```

```python
movies['keywords'] = movies['keywords'].apply(get_genres)
movies.head()
```

**'title'** Feature are good to go.
### Checking for "cast" feature

```python
movies.cast[0]
```

In this column, we have various attributes like cast_id, name_of_charachter, credit_id, gender, actor_name.
* In all attributes, only **actor_name** is of much use. So let's abstract **top-5 actors** for each movie

```python
def get_top5_actors(obj):
    Cast = []
    temp = 0
    for itr in literal_eval(obj):
        if temp < 5:
            Cast.append(itr['name'])
        temp+=1
    return Cast 
```

```python
movies['cast'] = movies['cast'].apply(get_top5_actors)
movies.head()
```

```python
print(movies.cast[0])
print(len(movies.cast[0]))
```

### Checking for "crew" feature

```python
movies.crew[0]
```

In this column, we have various attributes like credit_id, department, gender, job, name.


```python
def get_director(obj):
    Director  = []
    for itr in literal_eval(obj):
        if itr['job'] == 'Director':
            Director.append(itr['name'])
    return Director 
```

```python
movies['director'] = movies['crew'].apply(get_director)
movies.drop('crew', axis=1, inplace=True)
movies.head()
```

We **converge** genres,keywords,cast,directors into 1 column so we can work on 1 column only to find similarities that will help our model for better recommendations.

But before coverging, we have to perform some **additional preprocessing**.
* In **genres** , we have Science Fiction, which may be considered as 2 seperate identities science and fiction. So we have to convert Science Fiction into something like Science_Fiction or ScienceFiction
* In **cast/director** we have 2 different person with same first name --> Sam Worthington, Sam Mendes. So similarly we have to convert them into SamWorthington, SamMendes

```python
def converge(obj):
    ls = []
    for itr in obj:
        ls.append(itr.replace(" ",""))
    return ls
```

```python
movies['cast'] = movies['cast'].apply(converge)
movies['director'] = movies['director'].apply(converge)
movies['genres'] = movies['genres'].apply(converge)
movies['keywords'] = movies['keywords'].apply(converge)
```

```python
movies.head()
```

### Converging { genres,  keywords,  cast,  director } into 1 Column --> "features"

```python
movies['features'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['director']
movies.drop(['genres', 'keywords', 'cast', 'director'], axis=1, inplace=True)
movies.head()
```

```python
movies.features[0]
```

* Let's covert everything into 1 paragraph containing all important text; i.e. converting our list to string.

```python
movies['features'] = movies['features'].apply(lambda x: " ".join(x))
movies['features'][0]
```

* Convert everything in feature to small case to avoid any mis-match

```python
movies['features'] = movies['features'].apply(lambda x: x.lower())
movies['features'][0]
```

#### Our dataset seems good. Let's move to building our model 

## Building Recommender System

To build our Recommender System, we use **CountVectorizer()** function to get the similarities between movies by converting all movies into different vectors. Recommended movies are the movies that are closest to the vectors of our watched movie.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000)
```

```python
vector = cv.fit_transform(movies['features']).toarray()
```

```python
cv.get_feature_names()
```

To find Similarities between movies; we use **Cosine_similarity**. It finds shortest distance between vectors on the basis of angle.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
similarity.shape
```

* this matrix gives the similarity of all 4809 movies with each movie

Now define a function for recommendation which gives top5 movies based on similarity result

```python
def recommend(movie_name):
    index = movies[movies['title'] == movie_name].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for itr in distances[1:6]:
        print(movies.iloc[itr[0]].title)
```

### Let's Try our Recommender System

```python
recommend('The Avengers')
```

```python
recommend('Men in Black 3')
```

```python
recommend('Jurassic World')
```
